# Omdbapi Data Collection

```python
import os
import requests
import json
import pandas as pd
import numpy as np
from config import omdb_key,omdb_key2, omdb_key3, omdb_key4, omdb_key5, omdb_key6, omdb_key7
import pprint as pp
#each key has a limit of 1000 requests per day
keyList = [omdb_key, omdb_key2, omdb_key3, omdb_key4, omdb_key5, omdb_key6, omdb_key7]
```

## We used a csv file of offerings on netflix as our source of titles

```python
# Get movies from Netflix csv file that we downloaded from kaggle
movie_file = "Resources/netflix_titles.csv"
```

```python
nf_movies_df = pd.read_csv(movie_file)
#nf_movies_df.head()
```

## api Key Management
 - omdb allows a maximum of 1000 requests / day

```python
# will loop thru the list of api keys until all requests are used for the day
# apikeys are kept in an external config file and loaded into a list

url = "http://www.omdbapi.com/?t="
kl_index = 0
api_key = "&apikey=" + keyList[kl_index]
```

## Collect data from omdb database
- using titles from netflix, call the api and collect data of interest
- api response doesn't include boxoffice for listings that are not movies, so the code had to take that into account
- a counter is set to check how many requests have been used for a key, and the code moves to the next key if it is nearing the limit; this could have been handled by examining the response, but I wanted to leave some additional requests available / day for testing
- code handles titles that are not listed; 


```python
# loop thru the loaded netflix df, using titles to fetch info from API
#Boxoffice!!! some things don't have boxoffice!
max_count = np.arange(0,len(nf_movies_df))
extra_movie_info = []
NotAvailable = "N/A"
api_count = 0;

for i in max_count:
    genre = ""
    runtime = ""
    awards = ""
    rated = ""
    imdbRating = ""
    imdbVotes = ""
    poster = ""
    boxoffice = ""
    language = ""
    try: 
        response = requests.get(url + nf_movies_df['title'][i] + api_key)     
        data = response.json()
        #pp.pprint(data)
        if (data['Genre'] != NotAvailable):
            genre = data['Genre'] 
            
        if (data['Rated'] != NotAvailable):
            rated = data['Rated']
 
        if (data['Awards'] != NotAvailable):             
            awards = data['Awards']
            
        if (data['imdbRating'] != NotAvailable):
            imdbRating = data['imdbRating']
            
        if (data['imdbVotes'] != NotAvailable):
            imdbVotes = data['imdbVotes']

        if (data['Type'] == 'movie'):
            if (data['BoxOffice'] != NotAvailable):
                boxoffice = data['BoxOffice']

        if (data['Poster'] != NotAvailable):
            poster = data['Poster']
            
        if (data['Runtime'] != NotAvailable):
            runtime = data['Runtime']
            
        if (data['Language'] != NotAvailable):
            language = data['Language']
            
        movie_info = {
                "title": nf_movies_df['title'][i],
                "genre": genre,
                "runtime": runtime,
                "imdbRating": imdbRating,
                "imdbVotes": imdbVotes,
                "poster": poster,
                "awards": awards,
                "boxoffice": boxoffice,
                "language": language
                }
        
        extra_movie_info.append(movie_info) 
        
    except:
        print(f"No entry in omdb for {nf_movies_df['title'][i]}")
            
    finally:
        api_count = api_count + 1
        
    if (api_count > 950):
        print("Go to next Key")
        if (kl_index < (len(keyList)-1)):
            kl_index = kl_index + 1
            api_key = "&apikey=" + keyList[kl_index]
            api_count = 0
        else:
            print("We have used all the api key requests for today")
            break
    if (kl_index == len(keyList)):
        break
print(len(extra_movie_info))
```

## Saving data to a csv file
- while not necessary for final results, the code allows for appending to an existing file, so that testing could be interrupted and restarted

```python
#write the collected data to a csv file
omdb_data_list = pd.DataFrame(extra_movie_info)
omdb_data_list.head()
```

```python
# if file does not exist write header 
if not os.path.isfile('Resources\omdb_list.csv'):
    omdb_data_list.to_csv('Resources\omdb_list.csv', header='column_names')
else: # else it exists so append without writing the header
    omdb_data_list.to_csv('Resources\omdb_list.csv', mode='a', header=False)
```
